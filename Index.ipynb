```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pl
from mpl_toolkits.mplot3d import Axes3D

dat = pd.read_csv('all_combined.csv')

dat['my_wt']=dat['wow_bw_Nov']-dat['wow_bw_Aug']

s = dat['Sulfur']#
#        np.random.randint(0,50,size=305)
u = dat['Urea']#np.random.randint(0,50,size=305)
p = dat['P']#np.random.randint(0,50,size=305)
lwg = [x + np.abs(dat['my_wt'].min()) for x in dat['my_wt']]#np.random.sample(305)*20
lwg = [10*(x**0.5) for x in lwg]


fig = pl.figure()
ax = fig.add_subplot(111, projection='3d')

#for i in range(len(s)):
plot=ax.scatter(s,u,p,s=lwg,c=lwg,marker='o',cmap='viridis',alpha=0.5)
ax.set_xlabel('Sulphur Intake')
ax.set_ylabel('Phosphorus Intake')
ax.set_zlabel('Urea Intake')

pl.suptitle('Liveweight Gain with Supplement Intake')

cbar = fig.colorbar(plot)
cbar.set_label("Liveweight gain in kg")#, labelpad=-40)


pl.figure()
pl.subplot(1,3,1)
pl.plot(s,lwg,'ro')
pl.ylabel('LWG')
pl.xlabel('Sulfur')
pl.subplot(1,3,2)
pl.plot(p,lwg,'bo')
pl.ylabel('LWG')
pl.xlabel('Phos.')
pl.subplot(1,3,3)
pl.plot(u,lwg,'ko')
pl.ylabel('LWG')
pl.xlabel('Urea')


pl.figure()
pl.subplot(1,3,1)
ax=pl.gca()
plot=ax.scatter(s,p,s=lwg,c=lwg,marker='o',cmap='viridis',alpha=0.5)
pl.ylabel('Phos.')
pl.xlabel('Sulf')
pl.subplot(1,3,2)
ax=pl.gca()
plot=ax.scatter(u,p,s=lwg,c=lwg,marker='o',cmap='viridis',alpha=0.5)
pl.ylabel('Phos.')
pl.xlabel('Urea')
pl.subplot(1,3,3)
ax=pl.gca()
plot=ax.scatter(u,s,s=lwg,c=lwg,marker='o',cmap='viridis',alpha=0.5)
pl.ylabel('Sulf')
pl.xlabel('Urea')


pl.show()
```
